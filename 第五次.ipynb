```python
from datascience import *
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

table1 = Table.read_table("E:/hu/20211105-utf-8.csv")
table2 = Table.read_table("E:/hu/20211106-utf-8.csv")
table3 = Table.read_table("E:/hu/20211107-utf-8.csv")
table4 = Table.read_table("E:/hu/20211108-utf-8.csv")
table5 = Table.read_table("E:/hu/20211109-1-utf-8.csv")
table6 = Table.read_table("E:/hu/20211109-utf-8.csv")
table7 = Table.read_table("E:/hu/20211110-utf-8.csv")


name1 = input("请输入第一个名字：")
name2 = input("请输入第二个名字：")

def extract_rank(path,name):
    a = path.where('姓名',name).column(11).item(0)
    return a

rank1 = make_array(extract_rank(table1,name1),extract_rank(table2,name1),extract_rank(table3,name1),extract_rank(table4,name1),extract_rank(table5,name1),extract_rank(table6,name1),extract_rank(table7,name1))
results1 = Table().with_columns('考试序号',make_array(1,2,3,4,5,6,7),'班级名次',rank1)

rank2 = make_array(extract_rank(table1,name2),extract_rank(table2,name2),extract_rank(table3,name2),extract_rank(table4,name2),extract_rank(table5,name2),extract_rank(table6,name2),extract_rank(table7,name2))
results2 = Table().with_columns('考试序号',make_array(1,2,3,4,5,6,7),'班级名次',rank2)

make = make_array(1,2,3,4,5,6,7)
plt.plot(make,rank2,'*--',label=name2)
plt.plot(make,rank1,'o--',label=name1)
plt.title(name1 +'和'+ name2 + '比较')
plt.xlabel('历届考试')
plt.ylabel('年纪RANK')
plt.axis([0,10,650,150])
plt.legend(loc='upper right')
plt.rcParams['font.sans-serif']= ['Microsoft YaHei']
plt.grid()
```
